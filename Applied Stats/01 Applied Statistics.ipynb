2.An electrical manufacturing company conducts quality checks at specified periods on the products it manufactures. Historically, the 
failure rate for the manufactured item is 5%. Suppose a random sample of 10 manufactured items is selected. Answer the following 
questions.

```python
#Data given = N and Failure Rate . so Lets choose Binomial distribution.
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt 
n=10
p=0.05
q=1-p
k = np.arange(0,n+1)
Binomial_Distribution= stats.binom.pmf(k,n,p)
Binomial_Distribution_cdf= stats.binom.cdf(k,n,p)

#2.A. Probability that none of the items are defective?
print("Probability that none of the items are defective P(x=0)=",Binomial_Distribution[0].round(4))
```

```python
#2.B. Probability that exactly one of the items is defective?
print("Probability of exactly one of the items is defective P(x=1)=",Binomial_Distribution[1].round(4))
```

```python
#2.C. Probability that two or fewer of the items are defective? 
print("Using PMF- Probability that two or fewer of the items are defective p(x<=2)=",(Binomial_Distribution[0]+Binomial_Distribution[1]+Binomial_Distribution[2]).round(4))
print("Using CDF- Probability that two or fewer of the items are defective p(x<=2)=",(Binomial_Distribution_cdf[2]).round(4))
```

```python
#2.D. Probability that three or more of the items are defective ? 
prob_fewerthan3_defective=Binomial_Distribution[0]+Binomial_Distribution[1]+Binomial_Distribution[2] #p(x<3)

print("Using PMF -Probability that three or more of the items are defective p(x>=3)=",(1-prob_fewerthan3_defective).round(4)) #p(x>=3)=1-p(x<3)
print("Using CDF -Probability that three or more of the items are defective p(x>=3)=",(1-Binomial_Distribution_cdf[2]).round(4))
```

3.A car salesman sells on an average 3 cars per week. 

```python
# Avg rate is given , hence choose poisson distribution 
# lamba = 3 ( Avg number of succcess per week) , k =21 (Total number of success in a week is 7*3)
rate=3
k=np.arange(0,21)
Poisson_Distribution = stats.poisson.pmf(k,rate)
Poisson_Distribution_CDF = stats.poisson.cdf(k,rate)
#3.A. What is Probability that in a given week he will sell some cars? 
# he will sell some cars in the sense atleast one car which is greater than zero

print("Using - PMF Probability that in a given week he will sell some cars p(x>0)=",(1-Poisson_Distribution[0]).round(4)) #p(x>0)=1-p(x=0)
print("Using - CDF Probability that in a given week he will sell some cars p(x>0)=",(1-Poisson_Distribution_CDF[0]).round(4)) 
```

```python
#3.B. What is Probability that in a given week he will sell 2 or more but less than 5 cars?
print("Using - PMF Probability that in a given week he will sell 2 or more but less than 5 cars p(x>=2 and x<5)=",(Poisson_Distribution[2]+Poisson_Distribution[3]+Poisson_Distribution[4]).round(4)) #p(2,3,4)
print("Using - CDF Probability that in a given week he will sell 2 or more but less than 5 cars p(x>=2 and x<5)=",(Poisson_Distribution_CDF[4]-Poisson_Distribution_CDF[1]).round(4))
```

```python
#3.C. Plot the poisson distribution function for cumulative probability of cars sold per-week vs number of cars sold per week. 

plt.plot(k,Poisson_Distribution_CDF, 'o-')
plt.title('Poisson Distribution')
plt.xlabel('Number of Cars Sold per week')
plt.ylabel('Probability of Cars Sold per week')
plt.show()
```

4.Statistical Analysis:

```python
#4A. Is there any significant difference between Funds raised by companies that are still operating vs companies that closed down?

from statsmodels.stats.weightstats import ztest
Operatingstartup=Startupdata[(Startupdata['OperatingState']=='Operating')]
Closedstartup=Startupdata[(Startupdata['OperatingState']=='Closed')]
fig, ax = plt.subplots(1,2)
sns.distplot(Operatingstartup['Funds_in_million'], ax=ax[0])
sns.distplot(Closedstartup['Funds_in_million'], ax=ax[1])
plt.tight_layout()
```

```python
#4B. Write the null hypothesis and alternative hypothesis.
NullHypothesis = "There is no significant difference between fund raised by Operating company and Closed Company"
AlternateHypothesis = "There is  significant difference between fund raised by Operating company and Closed Company"

print("Null Hypothesis H0 = ",NullHypothesis)
print("Alternate Hypothesis Ha = ",AlternateHypothesis)
```

```python
#4C. Test for significance and conclusion
tstatistic, pval = ztest(Operatingstartup['Funds_in_million'],Closedstartup['Funds_in_million'])

if pval < 0.05:
    print(f'With a p-value of {round(pval,4)} the difference is significant. |We reject the null|')
else:
    print(f'With a p-value of {round(pval,4)} the difference is not significant. |We fail to reject the null|')
```

Conclusion : There is no significance difference between fund raised by Operating company and Closed Company
